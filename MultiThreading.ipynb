```python
Q1. What is multithreading in python? hy is it used? Name the module used to handle threads in python?
Ans1. Multithreading in Python is a technique of running multiple threads of execution simultaneously within a single process. Each thread runs concurrently and independently of each other, and they share the same memory space.
In Python, the threading module is used to handle threads. This module provides a simple and easy-to-use interface for creating and managing threads. It allows you to create new threads, start them, and communicate between them using various synchronization primitives such as locks, events, and semaphores.
```

```python
Q2. Why threading module used? write the use of the following functions
(i)activeCount
(ii)currentThread
(iii)enumerate
Ans2. The threading module in Python is used for creating and managing threads. It provides a high-level interface for creating and controlling threads, as well as synchronization primitives for inter-thread communication.
(i)activeCount: This function returns the number of thread objects that are currently active. An active thread is a thread that has been started and has not yet been terminated.
import threading

def my_func():
    print("This is my function")

threads = []
for i in range(5):
    t = threading.Thread(target=my_func)
    threads.append(t)
    t.start()

print("Active Threads:", threading.activeCount())

(ii) currentThread: This function returns a reference to the current thread object. This can be useful for identifying the current thread, and for passing the thread object to other functions.
    import threading

def my_func():
    print("Current thread is:", threading.currentThread().getName())

t = threading.Thread(target=my_func)
t.start()

(iii) enumerate: This function returns a list of all thread objects that are currently active. This can be useful for iterating over all the threads in a program, and for performing operations on them.
    import threading
import time

def my_func():
    time.sleep(1)

threads = []
for i in range(5):
    t = threading.Thread(target=my_func)
    threads.append(t)
    t.start()

for thread in threading.enumerate():
    print("Thread name:", thread.name)
```

```python
3. Explain the following functions
(i)run
(ii)start
(iii)join
(iv)isAlive
Ans3. (i) run: This method is called when the start method of a thread is called. It contains the code that will be executed in the new thread.
    import threading

class MyThread(threading.Thread):
    def run(self):
        print("This is my thread")

t = MyThread()
t.start()

(ii) start: This method creates a new thread and calls its run method. The code in the run method will be executed in the new thread.
    import threading

class MyThread(threading.Thread):
    def run(self):
        print("This is my thread")

t = MyThread()
t.start()

(iii) join: This method blocks the current thread until the thread on which it is called has finished executing. It can be useful for waiting for a thread to complete before continuing with the main thread.
    import threading
import time

class MyThread(threading.Thread):
    def run(self):
        time.sleep(2)
        print("This is my thread")

t = MyThread()
t.start()
print("Waiting for thread to finish")
t.join()
print("Thread finished")

(iv) isAlive: This method returns True if the thread on which it is called is still running, and False otherwise.
import threading
import time

class MyThread(threading.Thread):
    def run(self):
        time.sleep(2)
        print("This is my thread")

t = MyThread()
t.start()
print("Thread is running:", t.isAlive())
t.join()
print("Thread is running:", t.isAlive())
```

```python
Q4. Write a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes?
Ans4. import threading

def print_squares():
    for i in range(1, 11):
        print(f"Square of {i} is {i*i}")

def print_cubes():
    for i in range(1, 11):
        print(f"Cube of {i} is {i*i*i}")

t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
Q5. State advantages and disadvantages of multithreading?
Ans5. Advantages of Multithreading:
Improved Responsiveness: Multithreading allows a program to be more responsive to user inputs, since one thread can handle user input while another thread is running a lengthy task in the background.
Better Resource Utilization: By using multiple threads, a program can make better use of system resources like CPU and memory, resulting in improved performance.
    
Disadvantages of Multithreading:
Increased Complexity: Multithreading can make code more complex and harder to debug, as it can introduce issues like race conditions and deadlocks.
Increased Overhead: Multithreading comes with overhead, such as the cost of creating and switching between threads. In some cases, this overhead can outweigh the benefits of multithreading.
```

```python
Q6. Explain deadlocks and race conditions?
Ans6. Deadlock is a situation where two or more threads are blocked, waiting for each other to release a resource, which they currently hold. As a result, none of the threads can proceed further, and the program becomes stuck in an infinite loop. This can happen when two threads each hold a resource that the other thread needs to proceed.
A race condition is a situation where the output of a program depends on the sequence or timing of the execution of multiple threads. This can happen when two or more threads access a shared resource, such as a variable or a file, and the output of the program depends on the order in which these threads access the resource.
```
